```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import datetime
```

```python
df = pd.read_csv("connections.csv")
df.head()
```

```python
#empty directed graph
G2 = nx.DiGraph()
#add nodes and edges to graph
G2 = nx.from_pandas_edgelist(df, source = 'user_name', target='following',edge_attr='date_of_follow')
```

```python
#print nodes
G2.nodes()
```

```python
#set up layout
pos = nx.spring_layout(G2)
```

```python
# Convert Unix timestamp to a more manageable format (e.g., number of days since a specific date)
start_date = datetime.datetime(2023, 1, 1)  # Unix timestamp starts from this date
weights = list(nx.get_edge_attributes(G2, 'date_of_follow').values())
weights = [(datetime.datetime.fromtimestamp(i) - start_date).days for i in weights]

# Normalize weights so they are not exaggerated
weights = [i/50 for i in weights]
weights

```

```python
nx.draw_networkx_nodes(G2, pos, node_size=500, alpha=0.5)
nx.draw_networkx_edges(G2, pos,  edge_color='green')
nx.draw_networkx_labels(G2, pos)
plt.show()
```
